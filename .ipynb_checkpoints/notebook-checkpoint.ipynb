# Can you find a better way to segment your customers?

## üìñ Background
You work for a medical device manufacturer in Switzerland. Your company manufactures orthopedic devices and sells them worldwide. The company sells directly to individual doctors who use them on rehabilitation and physical therapy patients.

Historically, the sales and customer support departments have grouped doctors by geography. However, the region is not a good predictor of the number of purchases a doctor will make or their support needs.

Your team wants to use a data-centric approach to segmenting doctors to improve marketing, customer service, and product planning. 

## üíæ The data

The company stores the information you need in the following four tables. Some of the fields are anonymized to comply with privacy regulations.

#### Doctors contains information on doctors. Each row represents one doctor.
- "DoctorID" - is a unique identifier for each doctor.
- "Region" - the current geographical region of the doctor.
- "Category" - the type of doctor, either 'Specialist' or 'General Practitioner.'
- "Rank" - is an internal ranking system. It is an ordered variable: The highest level is Ambassadors, followed by Titanium Plus, Titanium, Platinum Plus, Platinum, Gold Plus, Gold, Silver Plus, and the lowest level is Silver.
- "Incidence rate"  and "R rate" - relate to the amount of re-work each doctor generates.
- "Satisfaction" - measures doctors' satisfaction with the company.
- "Experience" - relates to the doctor's experience with the company.
- "Purchases" - purchases over the last year.

#### Orders contains details on orders. Each row represents one order; a doctor can place multiple orders.
- "DoctorID" - doctor id (matches the other tables).
- "OrderID" - order identifier.
- "OrderNum" - order number.
- "Conditions A through J" - map the different settings of the devices in each order. Each order goes to an individual patient.

#### Complaints collects information on doctor complaints.
- "DoctorID" - doctor id (matches the other tables).
- "Complaint Type" - the company's classification of the complaints.
- "Qty" - number of complaints per complaint type per doctor.

#### Instructions has information on whether the doctor includes special instructions on their orders.
- "DoctorID" - doctor id (matches the other tables).
- "Instructions" - 'Yes' when the doctor includes special instructions, 'No' when they do not.

# **Can you find a relationship between purchases and complaints?**

The technique of preparing contingency tables for categorical variables is used.
A stacked bar graph is made to graphically analyze if there is a relationship between the Purchases variable and the Complaint Type. 
In order to find relationships between the two categorical variables, the data is organized in a contingency table. Occurrences are classified into groups in relation to the two variables. In the case in question, the Purchase variable is not categorical, so it must be discretized through the bins technique to transform it into categorical and be able to apply the contingency table technique.

**In the contingency table it can be seen that the totals between the two variables are quite similar, from which it could be thought that the variables are independent.**

## Contingence Table

**Convert Purchase to categorical variable**

The data necessary for the analysis is copied.

```python
datPurchase = CompPurchase[["Purchases","Complaint Type","Qty"]].copy()  
```

Quartiles are analyzed through the describe statement.
The Purchase variable is discretized into 4 bins.  

```python
datPurchase['Purchases'].describe()
```

You can see that 4 bins divide the data approximately proportionally, which corresponds to the quartiles.


```python
bins = [1,5,10,16,130]
names = ['muy bajo','medio', 'alto', 'muy alto']
datPurchase['PurchaseRange'] = pd.cut(datPurchase['Purchases'], bins, labels=names)
```

```python
datPurchase['PurchaseRange'].value_counts()
```

A contingency table is created by adding up the amount of purchase for each Complaint Type

```python
crossAnalize=pd.crosstab(datPurchase.PurchaseRange, datPurchase["Complaint Type"],  datPurchase["Qty"].fillna("0"), aggfunc='sum')
```

```python
crossAnalize
```

## Analysis using a bar chart

The figure shows the conditional distributions of purchase and complaints by means of a bar graph. As the bars are quite similar according to the complaints, it can be thought that there is no relationship between both variables. 
A chi square test will be performed to ensure such analysis.

```python
a = crossAnalize.plot(kind='bar', rot=0,stacked=False, colormap='tab10', figsize=(10, 6))
a.legend(title='Handedness', bbox_to_anchor=(1, 1.02),loc='upper left')

for rec in a.patches:
    height = rec.get_height()
    a.text(rec.get_x() + rec.get_width() / 2, 
              rec.get_y() + height / 2,
              "{:.0f}".format(height),
              ha='center', 
              va='bottom')

plt.suptitle('Analize relationship between purchases and complaints', fontsize=15, fontweight='bold')
a.set_title('it can be thought that there is no relationship between both variables', fontsize=12, fontweight='bold')

plt.show()
```

## **Chi Square Test**

In order to ensure that there is no relationship between both variables, the difference between the totals of the contingency table must not be statistically significant. To draw conclusions it is not possible to rely solely on the observation of the contingency table, a statistical test must be applied
Chi-Square Test- The test is applied when you have two categorical variables from a single population. It is used to determine whether there is a significant association between the two variables.
H‚ÇÄ (Null Hypothesis) ‚Äî that the 2 categorical variables being compared are independent of each other.
H‚ÇÅ (Alternate Hypothesis) ‚Äî that the 2 categorical variables being compared are dependent on each other.
If the p-value obtained is:

- less than 0.05 (the Œ± value you have chosen) you reject the H‚ÇÄ (Null Hypothesis) and accept the H‚ÇÅ (Alternate Hypothesis). This means the two categorical variables are dependent.
- greater than 0.05 you accept the H‚ÇÄ (Null Hypothesis) and reject the H‚ÇÅ (Alternate Hypothesis). This means the two categorical variables are independent.

**The chi square test was applied to the contingency table, obtaining a p-value of 0.224. From the p-value obtained, H0 is accepted, with** **which it can be ensured that the variables are independent , there is no relationship between Complaints and Purchases.**
**Independent**


```python
from scipy.stats import chi2_contingency

stat, p, dof, expected = chi2_contingency(crossAnalize)
  
# interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

## Doctor Clusterization

The Kmeans algorithm is applied. The goal of k-means is simple: it groups together similar data points with the goal of discovering underlying patterns. To achieve this goal, k-means searches for a fixed number (k) of groupings (clusters) in the data set.
The k-means algorithm identifies k number of centroids, and then assigns each data point to the closest group, while keeping the centroids as small as possible. 
The elbow or elbow curve is used to decide what is the optimal number of groups for the data in question.
In the graph of the Elbow curve we can see that the number of optimal cluster (k optimal) would be 3.

### Prepare data

The Satisfaction column is converted to numeric

```python
datFinal["Satisfaction"]=datFinal["Satisfaction"].replace('--',0)
```

```python
datFinal.shape
```

```python
datFinal["Satisfaction"]=datFinal["Satisfaction"].astype(float)
```

```python
datFinalK=datFinal
labelEncoder = LabelEncoder()
labelEncoder.fit(datFinalK['Category'])
datFinalK['Category'] = labelEncoder.transform(datFinalK['Category'])
```

```python
# Rank,Instructions,Region
labelEncoder = LabelEncoder()
labelEncoder.fit(datFinalK['Instructions'])
datFinalK['Instructions'] = labelEncoder.transform(datFinalK['Instructions'])

labelEncoder = LabelEncoder()
labelEncoder.fit(datFinalK['Region'])
datFinalK['Region'] = labelEncoder.transform(datFinalK['Region'])
```

```python
datFinalK.drop(columns=["DoctorID","Region"],axis='columns',inplace=True)
```

```python
datFinalK.isna().sum().sort_values()
```

```python
datFinalK["RankNro"]=datFinalK["RankNro"].replace(np.nan, 0)
```

```python
# se predice a que clase pertenece cada set de datos
# Se obtienen los centroides
X=datFinalK
from sklearn.preprocessing import  MinMaxScaler
scaler= MinMaxScaler()#instantiate
scaler.fit(X) # calcula la media y estandar para cada dimension
XNorm=scaler.transform(datFinalK)# transforma los datos a su nueva escala
XNorm=pd.DataFrame(XNorm,columns=datFinalK.columns)
```


### Best number of Clusters
Elbow curve

```python
# Run a number of tests, for 1, 2, ... num_clusters
num_clusters = 7
kmeans_tests = [KMeans(n_clusters=i, init='random', n_init=10) for i in range(1, num_clusters)]
score = [kmeans_tests[i].fit(XNorm).score(XNorm) for i in range(len(kmeans_tests))]

# Plot the curve
plt.figure(figsize=(12,6),dpi=80)
plt.plot(range(1, num_clusters),score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

### Application of k-means

Apply kmeans with k=3

```python
kmeans = KMeans(n_clusters = 3,                 # Set amount of clusters
                init = 'k-means++',             # Initialization method for kmeans
                max_iter = 300,                 # Maximum number of iterations
                n_init = 10,                    # Choose how often algorithm will run with different centroid
                random_state = 0)               # Choose random state for reproducibility

pred_y = kmeans.fit_predict(XNorm)
```

```python
XNorm["Cluster"]=pred_y
```

### Number of Doctors per group

```python
plt.suptitle('Number of Doctors per group', fontsize=15, fontweight='bold')

a = XNorm["Cluster"].value_counts().plot(kind='bar', rot=0,stacked=False, colormap='tab10', figsize=(10, 6))

for rec in a.patches:
    height = rec.get_height()
    a.text(rec.get_x() + rec.get_width() / 2, 
              rec.get_y() + height / 2,
              "{:.0f}".format(height),
              ha='center', 
              va='bottom')
```

# Identify which features impact the new segmentation strategy the most.

## CORRELATION MATRIX

First, we briefly we explain how to read the correlation matrix:

* Variables are not correlated if their value on the table where the variables intersect is near zero (bewteen -0.1 and +0.1).

* Highly correlated values are those where the value is furthest from zero (between -1 and -0.75 or between 0.75 and 1). Note, in particular, every variable is perfectly correlated to itself with value 1.

* Positive values between pairs of variables indicates that as one variable increase [resp. descreases] in value, the other variable increases [resp. decreases] in value.

* Negative values between pairs of variables indicates that as one variable increase [resp. descreases] in value, the other variable decreases [resp. increases] in value.

These matrices can be helpful for showing that certain variables are redundant when a pair of variables are highly correlated.




```python
plt.figure(figsize = (15,8))
sns.heatmap(XNorm.corr(), annot = True, cmap="Greens")
plt.suptitle('Correlation Matrix', fontsize=15, fontweight='bold')

```
